```python
import csv
from IPython.lib.pretty import pretty
import pprint
performance_csv = r'D:\repo\Stock\Seasonal-Stock\dataset\seasonal_top_performer.csv'

def get_top_performers_info_from_csv():
    top_performers = {}
    with open(performance_csv, mode='r') as csv_file:
        csv_reader = csv.DictReader(csv_file)
        for row in csv_reader:
            top_performer_info = {}
            for i in range(1,4):
                top_performer_info[i] = {'sym': row['top performer 1'], 'ratio':row['top performer 1 gain/loss ratio']}
            key = (row['month'], row['week'])
            top_performers[key] = top_performer_info
    return top_performers

pprint.pp(get_top_performers_info_from_csv())
```
